Cleaning the File

```python
df["Conversions"]=df["Conversions"].fillna(df["Conversions"].median())  #-- Filling in the Conversions Column
df["Clicks"]=df["Clicks"].fillna(df["Clicks"].median())                 #-- Filling in the first Clicks Column
```

```python
df["Channel"] = df["Channel"].fillna(df["Channel"].mode()[0])           #-- Filling in the Channel Column 
```

```python
df=df.drop_duplicates()                                                 #-- Dropping Duplicates
```

```python
df["Start_Date"]=df["Start_Date"].ffill()           #-- Filling in Null Date values
```

```python
print(df["Start_Date"].isna().sum())
```

```python
#-- Converting the datatypes

df["Campaign_Name"]=df["Campaign_Name"].astype("string")
df["Campaign_Tag"]=df["Campaign_Tag"].astype("string")

df["Start_Date"]=pd.to_datetime(df["Start_Date"], errors="coerce")
df["End_Date"]=pd.to_datetime(df["End_Date"], errors="coerce")

df["Channel"]=df["Channel"].astype("category")

#-- Don't know what to do with "Spend" Column.

df["Conversions"]=df["Conversions"].astype("Int64")

#df["Active"]=(df["Active"].astype(str).map({"true": True, "false": False, "yes": True, "no": False})) -- Don't know why it is not converting.
```

```python
df=df.loc[:,~df.columns.duplicated()]            #-- Removing the second CLICKS Column
df=df.drop(columns=["Clicks"])
```

```python
df.dtypes
```

```python
df.columns=df.columns.str.replace("_", " ")                   #-- Standardising Column Names
df.columns 
```

```python
import pandas as pd                                                      #Had to look up how to do this Ma'am.

# -----------------------------
# Load datasets
# -----------------------------
before = pd.read_csv("marketing_campaign.csv")
after = pd.read_csv("marketing_campaign_cleaned.csv")

# -----------------------------
# Basic Information
# -----------------------------
dataset_name = "marketing_campaign.csv"

rows_before = before.shape[0]
cols_before = before.shape[1]

rows_after = after.shape[0]
cols_after = after.shape[1]

# -----------------------------
# Missing Values
# -----------------------------
missing_before = before.isnull().sum()
missing_after = after.isnull().sum()

# -----------------------------
# Duplicate Rows
# -----------------------------
duplicates_before = before.duplicated().sum()
duplicates_after = after.duplicated().sum()

# -----------------------------
# Data Types
# -----------------------------
dtype_table = pd.DataFrame({
    "Column Name": after.columns,
    "Data Type": after.dtypes.astype(str)
})

# -----------------------------
# Numerical Summary
# -----------------------------
numeric = after.select_dtypes(include="number")

summary = pd.DataFrame({
    "Mean": numeric.mean(),
    "Median": numeric.median(),
    "Minimum": numeric.min(),
    "Maximum": numeric.max(),
    "Standard Deviation": numeric.std()
})

# -----------------------------
# Categorical Summary
# -----------------------------
categorical = after.select_dtypes(include=["object", "category"])

# -----------------------------
# Report
# -----------------------------

print("="*60)
print("DATA CLEANING SUMMARY REPORT")
print("="*60)

print(f"Dataset Name : {dataset_name}")
print(f"Total Rows : {rows_before}")
print(f"Total Columns : {cols_before}")

print("-"*50)
print("Missing Values")
print("-"*50)

print("Before Cleaning")
print(missing_before)

print("\nAfter Cleaning")
print(missing_after)

print("-"*50)
print("Duplicate Rows")
print("-"*50)

print(f"Duplicates Found : {duplicates_before}")
print(f"Duplicates Removed : {duplicates_before - duplicates_after}")

print("-"*50)
print("Data Types")
print("-"*50)

print(dtype_table.to_string(index=False))

print("-"*50)
print("Summary Statistics")
print("-"*50)

print(summary.round(2))

print("-"*50)
print("Categorical Summary")
print("-"*50)

for col in categorical.columns:
    print(f"\n{col}")
    print(f"Unique Values : {categorical[col].nunique()}")
    print(f"Most Frequent : {categorical[col].mode()[0]}")

print("-"*50)
print("Cleaning Actions")
print("-"*50)

actions = [
    "Removed duplicate rows",
    "Filled missing values",
    "Converted date columns to datetime",
    "Standardized column names",
    "Removed leading/trailing spaces",
    "Converted appropriate columns to categorical data types"
]

for action in actions:
    print(f"✔ {action}")

print("-"*50)
print("Final Dataset")
print("-"*50)

print(f"Rows : {rows_after}")
print(f"Columns : {cols_after}")

print("\nCleaning Completed Successfully")
print("="*60)
```
